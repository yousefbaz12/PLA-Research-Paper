```python
# Install required libraries if not already installed
# Run in terminal: pip install matplotlib matplotlib-venn

import matplotlib.pyplot as plt
from matplotlib_venn import venn3, venn3_circles

# Define the sets for Confidentiality, Integrity, and Availability based on Table 2
confidentiality = {'Eavesdropping', 'MITM', 'Adversarial ML Attacks'}
integrity = {'Spoofing', 'MITM', 'Sybil Attacks', 'Replay Attacks', 'Adversarial ML Attacks'}
availability = {'Jamming', 'Sybil Attacks'}

# Calculate subset sizes for Venn diagram (binary: 100, 010, 001, 110, 101, 011, 111)
subset_sizes = [
    len(confidentiality - integrity - availability),  # Only Confidentiality (100)
    len(integrity - confidentiality - availability),  # Only Integrity (010)
    len(availability - confidentiality - integrity),  # Only Availability (001)
    len(confidentiality & integrity - availability),  # Confidentiality & Integrity (110)
    len(confidentiality & availability - integrity),  # Confidentiality & Availability (101)
    len(integrity & availability - confidentiality),  # Integrity & Availability (011)
    len(confidentiality & integrity & availability)   # All three (111)
]

# Create the Venn diagram
plt.figure(figsize=(10, 8))
venn = venn3(subsets=subset_sizes, set_labels=('Confidentiality', 'Integrity', 'Availability'))

# Customize labels for each region with specific threats, checking for None
label_dict = {
    '100': 'Eavesdropping',
    '010': 'Spoofing\nReplay Attacks',
    '001': 'Jamming',
    '110': 'MITM\nAdversarial ML',
    '101': '',  # No threats here
    '011': 'Sybil Attacks',
    '111': ''   # No threats impact all three simultaneously
}

for region, text in label_dict.items():
    label = venn.get_label_by_id(region)
    if label is not None:  # Only set text if the label exists
        label.set_text(text)

# Style the diagram
for text in venn.set_labels:
    text.set_fontsize(14)
for label in venn.subset_labels:
    if label is not None:  # Avoid NoneType error for styling
        label.set_fontsize(12)

# Add title and adjust layout
plt.title("CIA Triad Impact of WIoT Security Threats", fontsize=16, pad=20)
plt.tight_layout()

# Save the diagram as an image file
plt.savefig('cia_triad_wiot_threats.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
from graphviz import Digraph

# Define the taxonomy as a nested dictionary
taxonomy = {
    "ML and DL Techniques for Physical-Layer Security in WIoT": {
        "Deep Learning Techniques": {
            "Autoencoders": {
                "Autoencoders for Anomaly Detection": {
                    "Technique Sophistication": "Intermediate",
                    "Applicability Context": "Healthcare IoT (e.g., detecting eavesdropping on wearable device data)",
                    "Threat Mitigated": "Eavesdropping"
                },
                "Autoencoders for Fingerprint Robustness": {
                    "Technique Sophistication": "Intermediate",
                    "Applicability Context": "Healthcare IoT (e.g., detecting hardware-agnostic tampering in medical devices)",
                    "Threat Mitigated": "Hardware-Agnostic Tampering"
                }
            },
            "CNN": {
                "CNN for RF Fingerprint Analysis": {
                    "Technique Sophistication": "Intermediate",
                    "Applicability Context": "Industrial IoT (e.g., detecting spoofing in machinery control systems)",
                    "Threat Mitigated": "Spoofing"
                },
                "CNN for Time-Frequency Analysis": {
                    "Technique Sophistication": "Intermediate",
                    "Applicability Context": "Smart Cities (e.g., detecting interference in traffic sensor networks)",
                    "Threat Mitigated": "Passive Jamming"
                }
            },
            "CVNN": {
                "CVNN for Complex Signal Processing": {
                    "Technique Sophistication": "Advanced",
                    "Applicability Context": "Smart Cities (e.g., detecting RF fingerprint tampering in access control systems)",
                    "Threat Mitigated": "RF Fingerprint Tampering"
                }
            },
            "GAN": {
                "Data Augmentation with GANs": {
                    "Technique Sophistication": "Advanced",
                    "Applicability Context": "Smart Cities (e.g., detecting hardware-agnostic tampering in diverse IoT devices)",
                    "Threat Mitigated": "Hardware-Agnostic Tampering"
                },
                "GAN for Synthetic Signal Generation": {
                    "Technique Sophistication": "Advanced",
                    "Applicability Context": "Smart Cities (e.g., detecting adaptive spoofing in traffic management systems)",
                    "Threat Mitigated": "Adaptive Spoofing"
                }
            },
            "RNN": {
                "LSTM": {
                    "Technique Sophistication": "Intermediate",
                    "Applicability Context": "Healthcare IoT (e.g., detecting MITM attacks on patient monitoring systems)",
                    "Threat Mitigated": "MITM"
                },
                "RNN with Attention": {
                    "Technique Sophistication": "Advanced",
                    "Applicability Context": "Industrial IoT (e.g., detecting cooperative eavesdropping in sensor networks)",
                    "Threat Mitigated": "Cooperative Eavesdropping"
                }
            },
            "Transformer Models": {
                "Transformers for Cross-Layer Analysis": {
                    "Technique Sophistication": "Advanced",
                    "Applicability Context": "Industrial IoT (e.g., detecting layered MITM in cross-layer systems)",
                    "Threat Mitigated": "Layered MITM"
                }
            }
        },
        "Machine Learning Techniques": {
            "Reinforcement Learning": {
                "DDPG": {
                    "Technique Sophistication": "Advanced",
                    "Applicability Context": "Intelligent Transportation Systems (e.g., detecting adaptive spoofing in V2V communication)",
                    "Threat Mitigated": "Adaptive Spoofing"
                },
                "Game-Theoretic RL": {
                    "Technique Sophistication": "Advanced",
                    "Applicability Context": "Smart Grid (e.g., detecting jamming of smart meter signals)",
                    "Threat Mitigated": "Active Jamming"
                },
                "Multi-Agent RL": {
                    "Technique Sophistication": "Advanced",
                    "Applicability Context": "Smart Cities (e.g., detecting cooperative eavesdropping on traffic sensors)",
                    "Threat Mitigated": "Cooperative Eavesdropping"
                },
                "Multi-Agent RL with Game Theory": {
                    "Technique Sophistication": "Advanced",
                    "Applicability Context": "Smart Cities (e.g., detecting layered MITM in multi-layer IoT systems)",
                    "Threat Mitigated": "Layered MITM"
                }
            },
            "Supervised Learning": {
                "Decision Tree": {
                    "Technique Sophistication": "Basic",
                    "Applicability Context": "Smart Grid (e.g., detecting MITM attacks on smart meter data)",
                    "Threat Mitigated": "MITM"
                },
                "Ensemble Learning": {
                    "Technique Sophistication": "Intermediate",
                    "Applicability Context": "Industrial IoT (e.g., detecting RF fingerprint tampering in device authentication)",
                    "Threat Mitigated": "RF Fingerprint Tampering"
                },
                "KNN": {
                    "Technique Sophistication": "Basic",
                    "Applicability Context": "Smart Cities (e.g., detecting spoofing of RFID tags)",
                    "Threat Mitigated": "Spoofing"
                },
                "SVM": {
                    "Technique Sophistication": "Basic",
                    "Applicability Context": "Industrial IoT (e.g., detecting jamming in factory sensor networks)",
                    "Threat Mitigated": "Active Jamming"
                }
            },
            "Unsupervised Learning": {
                "Clustering": {
                    "Technique Sophistication": "Basic",
                    "Applicability Context": "Dense IoT Deployments (e.g., factory floors with many sensors)",
                    "Threat Mitigated": "Passive Jamming"
                },
                "OCC-SVM": {
                    "Technique Sophistication": "Intermediate",
                    "Applicability Context": "Smart Homes (e.g., detecting eavesdropping on smart thermostat data)",
                    "Threat Mitigated": "Eavesdropping"
                }
            }
        }
    }
}

# Function to create a unique node ID (to avoid duplicates in Graphviz)
def create_node_id(path):
    return "_".join(path).replace(" ", "_").replace("(", "").replace(")", "").replace(".", "")

# Function to recursively build the Graphviz tree
def build_graphviz_tree(graph, node, parent_id=None, path=None):
    """
    Recursively builds a Graphviz tree from a nested dictionary.
    
    Args:
        graph (Digraph): The Graphviz Digraph object.
        node (dict or str): The current node in the taxonomy.
        parent_id (str): The ID of the parent node.
        path (list): The path from the root to the current node (used to create unique IDs).
    """
    if path is None:
        path = []

    if isinstance(node, dict):
        for key, value in node.items():
            # Create a unique node ID based on the path
            current_path = path + [key]
            node_id = create_node_id(current_path)
            
            # Add the node to the graph
            graph.node(node_id, label=key, shape="box" if isinstance(value, dict) else "ellipse")
            
            # Connect to the parent node if it exists
            if parent_id:
                graph.edge(parent_id, node_id)
            
            # Recursively build the subtree
            build_graphviz_tree(graph, value, node_id, current_path)
    else:
        # Leaf node (string value)
        leaf_id = create_node_id(path + [str(node)])
        graph.node(leaf_id, label=str(node), shape="ellipse")
        if parent_id:
            graph.edge(parent_id, leaf_id)

# Create a Graphviz Digraph object
dot = Digraph(comment="Taxonomy Tree", format="png")
dot.attr(rankdir="TB")  # Top to Bottom direction
dot.attr("node", shape="box", style="filled", fillcolor="lightblue")
dot.attr("edge", color="black")

# Build the Graphviz tree from the taxonomy
build_graphviz_tree(dot, taxonomy)

# Render the tree to a PNG file
output_file = "taxonomy_tree"
dot.render(output_file, view=False)  # Saves as taxonomy_tree.png

print(f"Taxonomy tree saved as {output_file}.png")
```

```python
fig, ax1 = plt.subplots()
ax1.plot(years, iot_devices, 'b-o', label='IoT Devices (Billions)', color='purple')
ax1.set_xlabel('Year')
ax1.set_ylabel('IoT Devices (Billions)', color='purple')
ax1.tick_params(axis='y', labelcolor='purple')
ax1.set_ylim(0, 35)
ax2 = ax1.twinx()
ax2.bar(years, security_papers, alpha=0.7, color='green', label='Security Papers (%)')
ax2.set_ylabel('Security Papers (% Growth)', color='green')
ax2.tick_params(axis='y', labelcolor='green')
ax2.set_ylim(0, 35)  # Adjust to match percentage scale
plt.title('Growth of IoT Devices and Security Papers (2020-2025)')
fig.legend(loc='upper left', bbox_to_anchor=(0.1,0.9))
for i, v in enumerate(security_papers):
    ax2.text(years[i], v + 0.5, f'{v}%', ha='center')
    plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Data for Physical-Layer Security Threats Taxonomy
threats_data = {
    "Threat": ["Active Jamming", "Passive Jamming", "Eavesdropping", "Cooperative Eavesdropping", 
               "Spoofing", "Adaptive Spoofing", "MITM", "Layered MITM", 
               "RF Fingerprint Tampering", "Hardware-Agnostic Tampering"],
    "Attack Vector": ["Signal Disruption", "Signal Disruption", "Signal Interception", "Signal Interception", 
                      "Signal Manipulation", "Signal Manipulation", "Signal Manipulation", "Signal Manipulation", 
                      "Hardware Exploitation", "Hardware Exploitation"],
    "Impact": ["Availability Disruption", "Availability Disruption", "Confidentiality Breach", "Confidentiality Breach", 
               "Integrity Violation", "Integrity Violation", "Confidentiality + Integrity", "Confidentiality + Integrity", 
               "Integrity Violation", "Integrity Violation"],
    "Attack Sophistication": ["Active", "Passive", "Passive", "Cooperative", 
                              "Active", "Cooperative", "Active", "Cooperative", 
                              "Active", "Cooperative"]
}

# Data for ML/DL Techniques Taxonomy
ml_dl_data = {
    "Technique": ["SVM", "KNN", "Decision Tree", "Ensemble Learning", "Clustering", "OCC-SVM", 
                  "Game-Theoretic RL", "Multi-Agent RL", "DDPG", "Multi-Agent RL with Game Theory", 
                  "CNN (Time-Frequency)", "CNN (RF Fingerprint)", "RNN with Attention", "LSTM", 
                  "GAN (Synthetic Signals)", "GAN (Data Augmentation)", "Autoencoders (Anomaly)", 
                  "Autoencoders (Fingerprint)", "Transformers", "CVNN"],
    "Threat Mitigated": ["Active Jamming", "Spoofing", "MITM", "RF Fingerprint Tampering", "Passive Jamming", "Eavesdropping", 
                         "Active Jamming", "Cooperative Eavesdropping", "Adaptive Spoofing", "Layered MITM", 
                         "Passive Jamming", "Spoofing", "Cooperative Eavesdropping", "MITM", 
                         "Adaptive Spoofing", "Hardware-Agnostic Tampering", "Eavesdropping", 
                         "Hardware-Agnostic Tampering", "Layered MITM", "RF Fingerprint Tampering"],
    "Technique Sophistication": ["Basic", "Basic", "Basic", "Intermediate", "Basic", "Intermediate", 
                                 "Advanced", "Advanced", "Advanced", "Advanced", 
                                 "Intermediate", "Intermediate", "Advanced", "Intermediate", 
                                 "Advanced", "Advanced", "Intermediate", "Intermediate", "Advanced", "Advanced"]
}

# Convert data to DataFrames
threats_df = pd.DataFrame(threats_data)
ml_dl_df = pd.DataFrame(ml_dl_data)

# Set up the plotting style
sns.set(style="whitegrid")

# Graph 1: Bar Chart of Threats by Attack Vector
plt.figure(figsize=(10, 6))
sns.countplot(data=threats_df, x="Attack Vector", palette="viridis")
plt.title("Distribution of Physical-Layer Security Threats by Attack Vector")
plt.xlabel("Attack Vector")
plt.ylabel("Number of Threats")
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig("threats_by_attack_vector.png")
plt.show()

# Graph 2: Pie Chart of Threats by Impact on CIA Triad
impact_counts = threats_df["Impact"].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(impact_counts, labels=impact_counts.index, autopct='%1.1f%%', colors=sns.color_palette("pastel"), startangle=140)
plt.title("Distribution of Threats by Impact on CIA Triad")
plt.tight_layout()
plt.savefig("threats_by_impact.png")
plt.show()

# Graph 3: Bar Chart of Threats by Attack Sophistication
plt.figure(figsize=(8, 6))
sns.countplot(data=threats_df, x="Attack Sophistication", palette="magma")
plt.title("Distribution of Threats by Attack Sophistication")
plt.xlabel("Attack Sophistication")
plt.ylabel("Number of Threats")
plt.tight_layout()
plt.savefig("threats_by_sophistication.png")
plt.show()

# Graph 4: Stacked Bar Chart of ML/DL Techniques by Threat Mitigated
plt.figure(figsize=(12, 8))
pivot_table = pd.crosstab(ml_dl_df["Threat Mitigated"], ml_dl_df["Technique Sophistication"])
pivot_table.plot(kind="bar", stacked=True, colormap="Set2", figsize=(12, 8))
plt.title("ML/DL Techniques by Threat Mitigated and Technique Sophistication")
plt.xlabel("Threat Mitigated")
plt.ylabel("Number of Techniques")
plt.xticks(rotation=45)
plt.legend(title="Technique Sophistication")
plt.tight_layout()
plt.savefig("ml_dl_by_threat.png")
plt.show()

# Graph 5: Bar Chart of ML/DL Techniques by Technique Sophistication
plt.figure(figsize=(8, 6))
sns.countplot(data=ml_dl_df, x="Technique Sophistication", palette="coolwarm")
plt.title("Distribution of ML/DL Techniques by Technique Sophistication")
plt.xlabel("Technique Sophistication")
plt.ylabel("Number of Techniques")
plt.tight_layout()
plt.savefig("ml_dl_by_sophistication.png")
plt.show()

print("Graphs saved as PNG files: threats_by_attack_vector.png, threats_by_impact.png, threats_by_sophistication.png, ml_dl_by_threat.png, ml_dl_by_sophistication.png")
```

```python
import matplotlib.pyplot as plt

# Data
years = [2020, 2021, 2022, 2023, 2024, 2025]
iot_devices = [10, 12, 15, 20, 25, 30]  # in billions
security_papers = [12.3, 15.8, 19.4, 23.7, 27.2, 31.5]  # in percentage

# Create a figure and axis with two y-axes
fig, ax1 = plt.subplots()

# Plot IoT devices on the left y-axis
ax1.plot(years, iot_devices, color='purple', marker='o', label='IoT Devices')
ax1.set_xlabel('Year')
ax1.set_ylabel('IoT Devices (billions)', color='purple')
ax1.tick_params(axis='y', labelcolor='purple')

# Create a second y-axis for security papers
ax2 = ax1.twinx()
ax2.bar(years, security_papers, color='lightgreen', alpha=0.6, label='Security Papers')
ax2.set_ylabel('Security Related Papers (%)', color='black')
ax2.tick_params(axis='y', labelcolor='black')

# Add legends
fig.legend(loc='upper left', bbox_to_anchor=(0.1,0.9))

# Add percentage labels on top of the bars
for i, v in enumerate(security_papers):
    ax2.text(years[i], v + 0.5, f'{v}%', ha='center', va='bottom')

# Set y-axis limits to match the graph
ax1.set_ylim(0, 35)
ax2.set_ylim(0, 160)

# Show the plot
plt.title('Growth of IoT Devices and Security-Related Papers (2020-2025)')
plt.show()
```
